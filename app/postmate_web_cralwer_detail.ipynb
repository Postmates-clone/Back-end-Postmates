```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import csv

url = 'https://postmates.com/'
driver = webdriver.Chrome('../chromedriver')
driver.get(url)

# input 객체 잡기
article = driver.find_element_by_class_name('geosuggest__input')
# key word 값 입력
keyword = '로스엔젤레스'
article.send_keys(keyword)
# 검색 버튼 클릭
driver.find_element_by_class_name('css-um3ari').click()



SCROLL_PAUSE_TIME = 2

#전체 창 높이 기록
last_height = driver.execute_script("return document.body.scrollHeight")

# 10 스크롤 내려서 요청
for i in range(10):
    # 스크롤 바닥까지 내리면서 인피니트 스크롤 이벤트 발생시킴
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 잠시 sleep
    time.sleep(SCROLL_PAUSE_TIME)

    # 스크롤 바닥에서 50 정도 올림
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight -50);")

    # 스크롤 바닥까지 내리면서 인피니트 스크롤 이벤트 발생시킴
    new_height = driver.execute_script("return document.body.scrollHeight")

    print('last_hight :  ' + str(last_height))
    print('new_height  :  ' + str(new_height))


    last_height = new_height


```

```python
keyword_list = ['San Francisco', 'New York', 'Los angeles' ]

# 셀레니움 이용해서 객체 잡기
div_css_26574z = driver.find_element_by_class_name('css-26574z')
div_nearby = div_css_26574z.find_elements_by_class_name('css-19f3h52')
    
san_francisco_image_list = list()
san_francisco_name_list = list()
san_francisco_url_list = list()
data = {}
for restaurant_div in div_nearby:
    a_tag = restaurant_div.find_element_by_tag_name('a')
    print(a_tag.get_attribute('href'))
    san_francisco_url_list.append(a_tag.get_attribute('href'))
    
    
    div_tag = restaurant_div.find_element_by_tag_name('div')    

    try:
        san_francisco_name_list.append(div_tag.find_element_by_tag_name('div').find_element_by_tag_name('div').get_attribute('title'))
        print(div_tag.find_element_by_tag_name('div').find_element_by_tag_name('div').get_attribute('title'))
    except:
        print('error')
        
    try:
        san_francisco_image_list.append(div_tag.find_element_by_tag_name('div').find_element_by_tag_name('img').get_attribute('src'))
#         print(div_tag.find_element_by_tag_name('div').find_element_by_tag_name('img').get_attribute('src'))
        
    except:
        print('eerrr1111')

for name, image, url in zip(san_francisco_name_list, san_francisco_image_list, san_francisco_url_list):
    rs_name = name
    data[rs_name] = url
    rs_image = 'image__'+ name 
    data[rs_image] = image
    rs_url =  'url__' + name
    data[rs_url] = url
```
